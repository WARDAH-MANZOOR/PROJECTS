```python

text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200, length_function=len)
doc = text_splitter.split_text(str(document))
# docs.append(i)
```

```python

embeddings = OpenAIEmbeddings(openai_api_key=OPENAI_API_KEY)
support_data = doc
```

```python
support_store = Chroma.from_texts(support_data, embeddings, collection_name="support",persist_directory="support_store")
support_store.persist()
```

```python
db = Chroma(persist_directory="support_store", embedding_function=embeddings)
```

```python
def retrieve_combined_documents(query, max_combined_docs=4):
    retriever = db.as_retriever(search_type="mmr")

    rev_doc = retriever.get_relevant_documents(query)
    lim_rev_doc = rev_doc[:max_combined_docs]

    docs = db.similarity_search(query)
    lim_docs = docs[:max_combined_docs]

    combined_docs = str(lim_rev_doc) + str(lim_docs)
    # combined_docs=db.similarity_search(query)

    return combined_docs
```

```python
import sqlite3
from gtts import gTTS
import os
from collections import Counter

# Connect to the database
conn = sqlite3.connect('generated_content_ratings.db')
c = conn.cursor()

# Create tables to store ratings for each language
c.execute('''CREATE TABLE IF NOT EXISTS English_Ratings
             (Title TEXT, Language TEXT, GeneratedContent TEXT, Explanation TEXT, Diversity REAL, HumanRating INTEGER)''')
c.execute('''CREATE TABLE IF NOT EXISTS Urdu_Ratings
             (Title TEXT, Language TEXT, GeneratedContent TEXT, Explanation TEXT, Diversity REAL, HumanRating INTEGER)''')
c.execute('''CREATE TABLE IF NOT EXISTS Sindhi_Ratings
             (Title TEXT, Language TEXT, GeneratedContent TEXT, Explanation TEXT, Diversity REAL, HumanRating INTEGER)''')
c.execute('''CREATE TABLE IF NOT EXISTS Pashto_Ratings
             (Title TEXT, Language TEXT, GeneratedContent TEXT, Explanation TEXT, Diversity REAL, HumanRating INTEGER)''')

# Function to insert rating into the respective table based on language
def insert_rating(table_name, title, language, generated_content, explanation, diversity, human_rating):
    c.execute(f"INSERT INTO {table_name} (Title, Language, GeneratedContent, Explanation, Diversity, HumanRating) VALUES (?, ?, ?, ?, ?, ?)",
              (title, language, generated_content, explanation, diversity, human_rating))
    conn.commit()

def text_to_speech(text, language_code, filename, sampling_rate=24000, format_='mp3'):
    tts = gTTS(text=text, lang=language_code)
    tts.save(filename)
    os.system(f"start {filename}")

def calculate_diversity(text):
    words = text.split()
    word_count = Counter(words)
    total_words = len(words)
    diversity = len(word_count) / total_words
    return diversity

def collect_human_judgments():
    print("Please rate the generated content on a scale of 1 to 5 (1 being the lowest and 5 being the highest):")
    rating = int(input("Rating: "))
    return rating

############## GPT PROMPT ####################
def gpt(prompt, title, language):
    inp = []
    systems = {"role": "system", "content": """
You're an Islamic Naat generation tool. You have to generate naat from a given title and also generate its description when asked in the local language of Pakistan.
              """}
    new_inp = inp
    rcd = retrieve_combined_documents(prompt)
    systems2 = {"role": "system", "content": str(rcd)}
    systems3 = {"role": "user", "content": "Language: " + language}

    systems4 = {"role": "system", "content": prompt + title}

    new_inp.insert(0, systems)
    new_inp.insert(1, systems2)
    new_inp.insert(2, systems3)
    new_inp.insert(3, systems4)

    completion = client.chat.completions.create(
        model="gpt-4-turbo",
        messages=new_inp,
        temperature=0.7,  # Set the temperature for creativity and coherence
        top_p=0.9         # Apply nucleus sampling with p=0.9
    )
    return completion

def final_process():
    lang = input("Please select language from the following: English, Urdu, Pashto, Sindhi: ").lower()
    while lang not in ['english', 'urdu', 'pashto', 'sindhi']:
        print("Invalid language selection. Please choose from: English, Urdu, Pashto, Sindhi.")
        lang = input("Please select language: ").lower()

    title = input("Please enter the title of the naat: ")
    rep_naat = gpt("Generate Naat from this title: ", title, lang).choices[0].message.content
    rep_naat = rep_naat.replace('*', '')

    desc_prompt = "Generate description of this naat: "
    desc_response = gpt(desc_prompt, rep_naat, lang).choices[0].message.content
    desc_response = desc_response.replace('*', '')


    # Print generated Naat and description
    print("Naat:\n" + rep_naat)
    print("\nDescription:\n" + desc_response)

    # Calculate diversity of generated content
    diversity = calculate_diversity(rep_naat)
    print(f"Diversity of generated content: {diversity}")

    # Collect human judgments
    human_rating = collect_human_judgments()

    # Insert rating into the database
    table_map = {
        'english': 'English_Ratings',
        'urdu': 'Urdu_Ratings',
        'pashto': 'Pashto_Ratings',
        'sindhi': 'Sindhi_Ratings'
    }
    table_name = table_map.get(lang)
    if table_name:
        insert_rating(table_name, title, lang, rep_naat, desc_response, diversity, human_rating)

    # Convert generated content to speech
    language_code = 'en' if lang == 'english' else 'ur'
    text_to_speech(rep_naat, language_code, 'generated_content.mp3', sampling_rate=24000, format_='mp3')
    text_to_speech(desc_response, language_code, 'explanation.mp3', sampling_rate=24000, format_='mp3')
    print("Audio files generated: 'generated_content.mp3' and 'explanation.mp3'")

    return rep_naat, desc_response

# Example usage For Sindhi
naat, description = final_process()
```
